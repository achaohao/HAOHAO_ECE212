### Exercise	2:	Area	of	a	Room	(Solved,	13	Lines)
Write a program that asks the user to enter the width and length of a room. Once these values have been read, your program should compute and display the area of the room. The length and the width will be entered as floating-point numbers. Include units in your prompt and output message; either feet or meters, depending on which unit you are more comfortable working with.

```python
#compute the area of the room

#read the dimensions from the user
length = float(input("Enter the length of the room in ft:"))
width = float(input("Enter the width of the room in ft:"))

#compute the area of the room
area = length * width

#display the results
print("The area of the room is", area, "squared feet")
```

```python
#two integer
r = float(input("Enter the radius of the circle in m:"))
pi= 3.14

area = (pi*r**2)
volume = ((4/3) * (pi * r ** 3))

print("The area of the circle is", area, "meter squared")
print("The volume of the circle is", volume, "meter cube")
```

```python
#compute the area of the circle

#read the dimensions from the user
r = float(input("Enter the radius of the circle in m:"))
pi = 3.14

#compute the area of the circle
area = pi*r**2
#compute the volume of the circle
volume = ((4/3) * (pi * r ** 3))

#display the results
print("The area of the circle is", area, "squared meter")
print("The volume of the circle is", volume, "cubic meter")
```
